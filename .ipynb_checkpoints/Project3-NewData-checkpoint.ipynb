```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LassoCV, RidgeCV, ElasticNet
import seaborn as sns
```

<h1>Build Working Panel</h1>

```python
files=os.listdir("/home/sufyan/Metis/Week4/Project3/data/")
all_dfs=[]
for file in files:
    all_dfs.append(pd.read_excel("/home/sufyan/Metis/Week4/Project3/data/"+file))
```

```python
m=[1,3,4,5,6,8,9,10,13,16,17]
```

```python
for i in range(len(all_dfs)):
    all_dfs[i]=all_dfs[i].rename(columns = { all_dfs[i].columns[0]:files[i]})
    all_dfs[i].set_index(files[i], inplace=True)
    all_dfs[i].columns = all_dfs[i].columns.astype(int)
    #all_dfs[i].columns = pd.to_datetime(all_dfs[i].columns)
```

```python
for i in range(len(all_dfs[1])):
    all_dfs[1].iloc[i].plot()
plt.show()
```

<h1>Create final non time-series df</h1>

```python
peacestatus=pd.read_csv('murder.csv')
peacestatus.set_index('Country', inplace=True)
peacestatus.fillna(0, inplace=True)
finallist=[]
cols=[]
finaldf=pd.DataFrame()
peaceyear='Rate'
for df in range(len(all_dfs)):
    for year in [2010,2012]:
        try: 
            finallist.append(all_dfs[df][year])
            cols.append('{}-{}'.format(files[df].split('.')[0], year))
        except: pass
finallist.append(peacestatus[peaceyear])
cols.append(peaceyear)
```

```python
finaldf=(pd.concat(finallist, axis=1))
finaldf.columns=cols
finaldf.dropna(inplace=True)
finaldf
```

```python
finaldf[peaceyear].plot(kind='hist', bins=2)
plt.show()
```

```python
sns.pairplot(finaldf, size =10, aspect = 1)
plt.show()
```

<h1>Binning</h1>

```python

finaldf['peace']=pd.qcut(finaldf[peaceyear], 2, labels=['0','1'])
#finaldf.drop(peaceyear, axis=1, inplace=True)

finaldf[0:100]
```

```python
pd.qcut(finaldf[peaceyear], 2, labels=['0','1'])
```

<h1>Modeling</h1>

```python
X_train, X_test, y_train, y_test = train_test_split(finaldf.iloc[:,0:-3], finaldf['peace'], test_size=.25)
model=KNeighborsClassifier(n_jobs=-1)
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_train_pred=model.predict(X_train)

print('test error: {}'.format(r2_score(y_test,y_pred)))
print('train error: {}'.format(r2_score(y_train, y_train_pred)))
#print(model.coef_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(finaldf.iloc[:,0:-3], finaldf['Rate'], test_size=.25)

lasso_est = Lasso(alpha=200)
lasso_est.fit(X_train, y_train)
y_pred = lasso_est.predict(X_test)

r2_score(y_test,y_pred)
```
